```python
from google.colab.patches import cv2_imshow
import cv2

# Load the color image
color_image = cv2.imread('/content/aliaa.jpg.crdownload')

# Resize the color image
resized_color_image = cv2.resize(color_image, (400, 400))

# Convert the color image to grayscale
gray_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)

# Resize the grayscale image
resized_gray_image = cv2.resize(gray_image, (400, 400))

# Display the resized original and grayscale images
cv2_imshow(resized_color_image)
cv2_imshow(resized_gray_image)
```
